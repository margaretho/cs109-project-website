## Linear Regression of Average Stars

```python
# For Elite Users and "Review Count, Time as Yelper, and Friend Count"
elites_index = np.random.rand(len(elites)) < 0.8
elites_train = elites[elites_index]
elites_test = elites[~elites_index]
X_elites_train = elites_train[['review_count','timeInDays','friend_count']].values
Y_elites_train = elites_train[['average_stars']].values
X_elites_test = elites_test[['review_count','timeInDays','friend_count']].values
Y_elites_test = elites_test[['average_stars']].values

alpha = 0.1
elites_lasso = Lasso(alpha=alpha)
elites_lasso_fit = elites_lasso.fit(X_elites_train, Y_elites_train)
elites_y_pred_lasso = elites_lasso_fit.predict(X_elites_test)
print "For Elites, the regression coefficients for review count, time as a Yelper, and the friend count are", elites_lasso_fit.coef_
print "For Elites, the intercept for the regression is", elites_lasso_fit.intercept_
```

```python
# For Normal Users and "Review Count, Time as Yelper, and Friend Count"
normals_index = np.random.rand(len(normals)) < 0.8
normals_train = normals[normals_index]
normals_test = normals[~normals_index]
X_normals_train = normals_train[['review_count','timeInDays','friend_count']].values
Y_normals_train = normals_train[['average_stars']].values
X_normals_test = normals_test[['review_count','timeInDays','friend_count']].values
Y_normals_test = normals_test[['average_stars']].values

normals_lasso = Lasso(alpha=alpha)
normals_lasso_fit = normals_lasso.fit(X_normals_train, Y_normals_train)
normals_y_pred_lasso = normals_lasso_fit.predict(X_normals_test)
print "For Normals, the regression coefficients for review count, time as a Yelper, and the friend count are", normals_lasso_fit.coef_
print "For Normals, the intercept for the regression is", normals_lasso_fit.intercept_
```

```python
normal_reg_data = normals.drop(['yelping_since','elite','type','friends','name','user_id','yelping_since_datetime'], 1)
elite_reg_data = elites.drop(['yelping_since','elite','type','friends','name','user_id','yelping_since_datetime'], 1)
reg_list = normal_reg_data.columns.values.tolist()
reg_list.remove('average_stars')

# For Elite Users - All Data
elite_reg_data_index = np.random.rand(len(elite_reg_data)) < 0.8
elite_reg_data_train = elite_reg_data[elite_reg_data_index]
elite_reg_data_test = elite_reg_data[~elite_reg_data_index]
X_elite_reg_data_train = elite_reg_data_train[reg_list].values
Y_elite_reg_data_train = elite_reg_data_train[['average_stars']].values
X_elite_reg_data_test = elite_reg_data_test[reg_list].values
Y_elite_reg_data_test = elite_reg_data_test[['average_stars']].values

alpha = 0.3
elite_reg_data_lasso = Lasso(alpha=alpha)
elite_reg_data_lasso_fit = elite_reg_data_lasso.fit(X_elite_reg_data_train, Y_elite_reg_data_train)
elite_reg_data_y_pred_lasso = elite_reg_data_lasso_fit.predict(X_elite_reg_data_test)
elite_coeficients = pd.DataFrame(reg_list,elite_reg_data_lasso_fit.coef_)
print "For Elites, the regression coefficients for", elite_coeficients
print "For Elites, the intercept for the regression is", elite_reg_data_lasso_fit.intercept_
print("Mean squared error for elites: %.2f"
      % np.mean((Y_elite_reg_data_test - elite_reg_data_y_pred_lasso) ** 2))

# For Normal Users - All Data
normal_reg_data_index = np.random.rand(len(normal_reg_data)) < 0.8
normal_reg_data_train = normal_reg_data[normal_reg_data_index]
normal_reg_data_test = normal_reg_data[~normal_reg_data_index]
X_normal_reg_data_train = normal_reg_data_train[reg_list].values
Y_normal_reg_data_train = normal_reg_data_train[['average_stars']].values
X_normal_reg_data_test = normal_reg_data_test[reg_list].values
Y_normal_reg_data_test = normal_reg_data_test[['average_stars']].values

normal_reg_data_lasso = Lasso(alpha=alpha)
normal_reg_data_lasso_fit = normal_reg_data_lasso.fit(X_normal_reg_data_train, Y_normal_reg_data_train)
normal_reg_data_y_pred_lasso = normal_reg_data_lasso_fit.predict(X_normal_reg_data_test)
normal_coeficients = pd.DataFrame(reg_list,normal_reg_data_lasso_fit.coef_)
print "For Normals, the regression coefficients for", normal_coeficients
print "For Normals, the intercept for the regression is", normal_reg_data_lasso_fit.intercept_
```

## Gender Analysis

We proceeded then to do a gender analysis to see if there are any differences between the genders. To determine the gender of the user (as this was not provided by Yelp in its dataset), we looked at the user names provided and used census data to determine if a name was generally female or male.  We then assigned genders to each user based on this, so that we could explore differences.

```python
# Obtained from https://www.census.gov/topics/population/genealogy/data/1990_census/1990_census_namefiles.html via https://raw.githubusercontent.com/Bemmu/gender-from-name/master/gender.py
file = open('gender.txt')
data = file.read()
data = data[3:]
gender = ast.literal_eval(data)

# Import and use gender data
genderdf = pd.DataFrame.from_dict(gender)
genderdf = genderdf.T
def applygender(x):
    if x[0]!=x[1]:
        return 'uncertain'
    else: 
        return x[0]
genderdf = pd.DataFrame(genderdf.apply(applygender,axis=1))
genderdf = genderdf.reset_index()
names = genderdf.apply(lambda x: x['index'].capitalize(),axis=1)
genderdf['names'] = names
genderdf = genderdf.drop('index', axis=1)
genderdf = genderdf.rename(columns = {0:'gender'})

# Add genders to elites
elitegenders = pd.DataFrame.merge(elites,genderdf,left_on='name',right_on='names',how='left')
elitegenders = elitegenders.drop('names',axis=1)
elitegenders[['gender']] = elitegenders[['gender']].fillna('uncertain')
elitemale = elitegenders.gender.str.contains('male').sum()
elitefemale = elitegenders.gender.str.contains('female').sum()
eliteuncertain = elitegenders.gender.str.contains('uncertain').sum()

# Add genders to normals
normalgenders = pd.DataFrame.merge(normals,genderdf,left_on='name',right_on='names',how='left')
normalgenders = normalgenders.drop('names',axis=1)
normalgenders[['gender']] = normalgenders[['gender']].fillna('uncertain')
normalmale = normalgenders.gender.str.contains('male').sum()
normalfemale = normalgenders.gender.str.contains('female').sum()
normaluncertain = normalgenders.gender.str.contains('uncertain').sum()
```

### Looking at the 500 most useful terms and two- and three-word phrases, there was a clear trend in language used to describe or discuss food and language used to describe and discuss service.  Many of the most frequently used, most useful one-word terms made appearances in the phrases, so we chose to use the keywords with clear food or atmosphere association to proceed with the sentiment analysis.

```python
# vocab for sentiment analysis, divided into food-related and feel(atmosphere)-related
vocab = pd.read_csv('yelp_dataset/vocab.csv')
vocab.head()
```

```python
# to see if there is a difference in tone for the reviews, whether more food-related or atmosphere-related
usefullen = len(useful)
foodcounts = []
feelcounts = []
for i in xrange(usefullen):
    foodcount = 0
    feelcount = 0
    split = str.split(useful.iloc[i]['text'])
    for word in split:
        if word in vocab['food'].values:
            foodcount += 1
        elif word in vocab['feel'].values:
            feelcount += 1
        else:
            continue
    foodcounts.append(foodcount)
    feelcounts.append(feelcount)

# positive values mean more food-related, negative means more atmosphere-related
diff = [a - b for a, b in zip(foodcounts, feelcounts)]
```

```python
# plot the differences
plt.hist(diff, bins = 50)
plt.title("Differences Between Number of Food-Related Terms Used in a Review vs. Atmosphere-Related Terms")
plt.ylabel("Frequency")
plt.xlabel("Difference")
plt.show()
```

### Looking at all of the useful reviews, there is a range in tone with some reviews talking much more about food and some reviews talking much more about the atmosphere, and some speaking more or less equally about both.

```python
# add differences column
useful['diff'] = diff

# separate reviews into food or atmosphere reviews
foodreviews = useful[useful['diff'] > 0]
feelreviews = useful[useful['diff'] < 0]

# average of star rating for food-related reviews
foodscores = foodreviews.groupby('business_id', as_index = False).mean()
foodscores.rename(columns={'stars':'foodscore'}, inplace=True)

# compare overall restaurant star rating with food rating
restaurantfood = pd.merge(restaurant[['business_id','stars']], foodscores[['business_id','foodscore']], on = ['business_id'])

plt.scatter(restaurantfood['foodscore'],restaurantfood['stars'])
plt.title("Restaurant Rating vs. Food Score")
plt.ylabel("Stars")
plt.xlabel("Food Score")
plt.show()
```

```python
# average of star rating for atmosphere-related reviews
feelscores = feelreviews.groupby('business_id', as_index = False).mean()
feelscores.rename(columns={'stars':'feelscore'}, inplace=True)

# compare overall restaurant star rating with atmosphere rating
restaurantfeel = pd.merge(restaurant[['business_id','stars']], feelscores[['business_id','feelscore']], on = ['business_id'])

plt.scatter(restaurantfeel['feelscore'],restaurantfeel['stars'])
plt.title("Restaurant Rating vs. Atmosphere Score")
plt.ylabel("Stars")
plt.xlabel("Atmosphere Score")
plt.show()
```

```python
# compare all three scores
allscores = pd.merge(restaurantfood[['business_id','stars','foodscore']], restaurantfeel[['business_id','feelscore']], on = ['business_id'])
allscores.head()

x = sns.lmplot("foodscore", "feelscore", hue="stars", data=allscores, fit_reg=False)
x.set_ylabels("Atmosphere Score")
x.set_xlabels("Food Score")
plt.title("Atmosphere Score vs. Food Score for Overall Restaurant Stars")
plt.show()
```

### Generally, restaurants have similar food and atmosphere scores, which relate to their overall score, but there are some more extreme cases with very high food scores and very low atmosphere scores, which is more common than high atmosphere scores and low food scores, though that does occur in a few cases.

```python
# see if there's any difference if we limit the analysis to restaurants with more than 5 food reviews or atmosphere reviews
allids = foodreviews['business_id'].values
morethanfivefoodid = [business for business in allids if len(foodreviews[foodreviews['business_id'] == business]) >= 5]
morethanfivefood = restaurantfood[restaurantfood['business_id'].isin(morethanfivefoodid)]
morethanfivefeelid = [business for business in allids if len(feelreviews[feelreviews['business_id'] == business]) >= 5]
morethanfivefeel = restaurantfeel[restaurantfeel['business_id'].isin(morethanfivefeelid)]

plt.scatter(morethanfivefood['foodscore'],morethanfivefood['stars'])
plt.title("Restaurant Rating vs. Food Score for Restaurants with At Least 5 Food Scores")
plt.ylabel("Stars")
plt.xlabel("Food Score")
plt.show()

plt.scatter(morethanfivefeel['feelscore'],morethanfivefeel['stars'])
plt.title("Restaurant Rating vs. Atmosphere Score for Restaurants with At Least 5 Atmosphere Scores")
plt.ylabel("Stars")
plt.xlabel("Atmosphere Score")
plt.show()

# more than 5 reviews for both food and atmosphere
morethanfivetotalid = [business for business in allids if len(foodreviews[foodreviews['business_id'] == business]) >= 5 and len(feelreviews[feelreviews['business_id'] == business]) >= 5]
morethanfivetotal = allscores[allscores['business_id'].isin(morethanfivetotalid)]

x = sns.lmplot("foodscore", "feelscore", hue="stars", data=morethanfivetotal, fit_reg=False)
x.set_ylabels("Atmosphere Score")
x.set_xlabels("Food Score")
plt.title("Atmosphere Score vs. Food Score for Overall Restaurant Stars with At Least 5 Reviews for Each Category")
plt.show()
```

### When we restrict the data to only restaurants with at least 5 reviews for each category of food and atmosphere, many of the outliers, where the food score is high and the atmosphere score low, disappear.  This means that highly rated restaurants generally have good food and a good atmosphere, and few people express highly polar opinions.  It must be noted, however, that the sentiment analysis is quite naive and may miss a lot of the nuances of the review language.
